## Experimento transfer learning

```python
# Using numpy we can use the function loadtxt to load your CSV file.
# We ignore the first line with the column names and use ',' as a delimiter.
data = np.loadtxt('experimento_transfer_learning.csv', delimiter=',', skiprows=1)

['Epoch', 'mace', 'val_mace']

epoch = data[:,0]
mace = data[:,1]
val_mace = data[:,2]

fig, ax = plt.subplots(figsize=(10,10))

# train
ax.plot(epoch, mace, 'green',label='Training MACE')
ax.plot(epoch, val_mace, 'red',label='Validation MACE')

# Show the legend
plt.legend()
```
